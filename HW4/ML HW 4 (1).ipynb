```python
import itertools
def comb(dup,k):
    combeg=[]
    for ab in itertools.combinations(dup, k):
        combeg+= [list(ab)]
    return combeg 
def countfreq(ad):
    count1=0
    for row in ilist:
        d=1
        for b in ad:
            if b not in row:
                d=0
                break
        if d !=0:
            count1+=1
    return count1
def unique(ilist):
    distinct1=[]
    for sublist in ilist:
        for ele in sublist:
            if ele not in distinct1:
                distinct1.append(ele)
                
    distinct1.sort()            
    return distinct1
def most_freq(tplist,combs):
    freqL=[]
    #print(combs)
    for i in combs:
        for j in tplist:
            #print(i[:len(i)-1])
            if i[:len(i)-1] == j:
                freqL.append(i)
    return freqL
```

## Fk-1 x Fk-1

```python
def combfk_1(dup,k):
    combeg=[]
    for ab in itertools.combinations(dup, k):
        combeg+= [list(ab)]
    return combeg
def merge(f3,k):
    f3 = list(f3)
    f3.sort()
    mitems = []
    for i in range(len(f3)-1):
        for j in range(i+1,len(f3)):
            if f3[i][:k-1] == f3[j][:k-1]:
                mitems.append(list(set(f3[i]+f3[j])))
    return mitems
```

```python
mincnt = 3
support = mincnt*100/len(items1)
k = 1
templist=[]
c = mincnt
ilist = list(items1)
uniq1 = unique(ilist)
max_k = len(uniq1)
prune = []
while c>=mincnt and k <= max_k:
    combs=[]
    clist=[]
    freqL=[]
    templist.append(k-1)
    templist[k-1]=[ ]
    combs = combfk_1(uniq1,k)
    if len(combs)==0:
        break
    print("Step",k,":")
    for ad in combs:
        count1=countfreq1(ad)
        print(ad," ",count1)
        clist.append(count1)
        if count1>=mincnt:
            templist[k-1].append(ad)
    c=max(clist)
    if c>=mincnt:
        nwtp=templist[k-1]
        print("The list with min support",support,"% is ",templist[k-1])   
    if k>=2:
        #print('k-1', templist[k-1])
        merge1= merge(templist[k-1],k)
        uniq1=unique1(merge1)
    else:
        uniq1 = unique1(templist[k-1])
    k+=1
```

```python
conf = 0.67
print("Frequent item list with min confidence",conf*100,"%")
#print(nwtp)
for ad in nwtp:
    temp2=[]
    for i in range(len(ad),0,-1):
        temp2=comb(ad,i-1)
        for af in temp2:
            confid=float(countfreq(ad)/countfreq(af))
            if confid>=conf:
                x=list((set(ad)-set(af)))
                print(af,"=>",x)
```

## Analysis regarding both approaches

In case of Fk-1 x F1 we can observe in case of the small dataset that it generated 4 itemsets of 3 items
but in case of Fk-1 x Fk-1 we can observe for same data only 1 three itemset is generated at the final step.
    This is because Fk-1 x Fk-1 approach we check atleast previous 2 frequent itemsets. Therefore it will result in reduction in number of itemsets for larger datasets
